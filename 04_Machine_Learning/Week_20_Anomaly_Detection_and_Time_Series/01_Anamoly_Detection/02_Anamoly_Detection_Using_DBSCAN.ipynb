# <font color = 'orange'> Anamoly Detection using DBSCAN

---

### <font color = 'Blue'> Import Library, load Dataset

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
# noise is added to spread the data points
x, y = make_circles(n_samples = 750, factor = 0.3, noise = 0.1)
```

### <font color = '#AA00FF'> Observation :
* It is unsupervised learning, so we will not use y i.e output feature.

```python
x
```

---

### <font color = 'Blue'> Plot input features

```python
plt.scatter(x[:,0],x[:,1])
```

### <font color = '#AA00FF'> Observation :
* we can notice 2 groups.

---

### <font color = 'Blue'> Clustering

```python
dbscan = DBSCAN(eps = 0.10)
```

```python
dbscan.fit(x)
```

```python
dbscan.labels_
```

### <font color = '#AA00FF'> Observation :
* we can notice that there are multiple cluster groups and -1 value i.e there are some outliers/noise.

---

### <font color = 'Blue'> Plot input features with respect to predict labels

```python
plt.scatter(x[:,0], x[:,1], c = dbscan.labels_)
```

### <font color = '#AA00FF'> Observation :
* we can notice that there are multiple cluster groups because of noise not only 2 cluster groups.

---

### <font color = 'Blue'> Plot input features with respect to actual output feature which we should not use

```python
plt.scatter(x[:,0], x[:,1], c = y)
```
