### <font color = 'Blue'> Import Library, load Dataset

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
```

```python
x, y = make_circles(n_samples = 750, noise = 0.12, factor = 0.3)
```

```python
x
```

---

```python
plt.scatter(x[:,0],x[:,1])
```

---

```python
from sklearn.neighbors import LocalOutlierFactor

# p - distance metrics and 2 value denotes eucledian distance
lof = LocalOutlierFactor(n_neighbors = 7, algorithm = 'ball_tree', p = 2)
```

```python
lof.fit(x)
```

```python
y_pred = lof.fit_predict(x)

# -1 denotes the outliers
y_pred
```

```python
plt.scatter(x[:,0], x[:,1], c = y_pred)
```

### <font color = '#AA00FF'> Observation :
* We can notice the outliers they are detected based on the density of the point with the neares neighbors.
