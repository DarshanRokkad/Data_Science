# <font color = 'orange'> Feature Scaling 

### There are 3 types of feature scaling 
1. Standardization.
2. Normalization - MinMaxScaler.
3. Unit Vector.

---

## <font color='blue'> 1. Standardization.

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

### To do standardization we have to apply z score for feature

```python
import numpy as np 

mean = np.mean(df['total_bill'])
std = np.std(df['total_bill'])
```

```python
print('Mean :',mean)
print('Standard deviation :',std)
```

### Apply z score for each and every value 
z-score = (x - mean) / standard_deviation

```python
normalised_data = []
for x in list(df['total_bill']):
    z_score = (x-mean)/std
    normalised_data.append(z_score)
    
# data which is scaled downed to standard scale after applying z score
normalised_data
```

### Distribution will not be changed only the value will be scaled down to standard scale only  

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(normalised_data)
```

---

### We can do standardization using sklearn 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
# scaler is object of StandardScaler class
scaler
```

```python
scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```

---

```python
# we have another method for fit and transfer
scaler.fit_transform(df[['total_bill']])
```

```python
# Creating dataframe
import pandas as pd

new_df = pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]),columns=['standardized_total_bill','tip'])
new_df
```

---

```python
# we can just apply to get standardized values for new data like below
scaler.transform([[13,4]])
```

## <font color='blue'> 2.Normalization - MinMaxScaler

```python
df=sns.load_dataset('diamonds')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()

min_max
```

```python
min_max.fit_transform(df[['depth','table','price']])
```

```python
# for any new data scaling can be done
min_max.transform([[65,57,325]])
```
