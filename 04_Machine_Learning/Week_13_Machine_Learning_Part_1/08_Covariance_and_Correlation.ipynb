# <font color = 'orange'> Covariance and Correlation

---

```python
import seaborn as sns
```

```python
df = sns.load_dataset('healthexp')

df
```

---

## <font color='blue'> Covariance

```python
import numpy as np
```

```python
df.cov()
# relation between Spending_USD and Life_Expectancy
```

---

## <font color='blue'> Correlation 
    1.Pearson correlation coefficient.
    2.Spearman rank correlation.

```python
# by default it will be pearson
df.corr(method = 'pearson')
```

```python
df.corr(method = 'spearman')
```

## <font color='blue'> Other Example

```python
df = sns.load_dataset('penguins')

df
```

```python
df[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']].var()
```

```python
df.corr()

# by this we can conclude that when bill_depth_mm increasing flipper_length_mm decreasing
```

```python
df.corr(method='spearman')
```
