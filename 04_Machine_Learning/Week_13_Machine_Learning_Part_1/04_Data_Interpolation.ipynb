# <font color = 'orange'> Data interpolation

Data interpolation is the process of estimating unknown values within a dataset based on the known values.  
In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas.  

Here is an example of how to perform data interpolation using the NumPy library:
1. Linear Interpolation
2. Cubic Interpolation With Scipy
3. Polynomial Interpolation

---

## 1. Linear interpolation

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2*i for i in x])
```

```python
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
# we can see that x and y are linear 
```

```python
# Now we want more data points so we can add to the dataset 
# interpolate the data using linear interpolation

x_new = np.linspace(1,5,10)  # create new x values
y_interp = np.interp(x_new,x,y) # creating unknown y datapoints using known x and y 
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
# we can see that more datapoints are add within that dataset
```

## 2. Cubic interpolation with scipy

```python
x = np.array([1,2,3,4,5])
y = np.array([i**3 for i in x])
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y)
```

```python
# creating interpolation in 1 dimension
from scipy.interpolate import interp1d
```

```python
# we are creating a function
f = interp1d(x,y,kind='cubic')
type(f)
```

```python
x_new = np.linspace(1,5,10)
y_interp = f(x_new)
```

```python
print(x_new)
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
# now we can see that more datapoints has been created 
```

## 3. Polynomial interpolation 

```python
x = np.array([1,2,3,4,5])
y = np.array([i**2 for i in x])
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y)
```

```python
# interpolate the data using polynomail interpolation

poly_function = np.polyfit(x,y,2)
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.polyval(poly_function,x_new)
```

```python
plt.scatter(x_new,y_interp)
```
