### <font color = 'Blue'> 1. Independent and dependent feature

```python
# independent feature 
x = dataset.iloc[:,:-1]
# df.iloc[row,columns]

# dependent feaure
y = dataset.iloc[:,-1]
```

```python
x.head(2)
```

```python
y.head(2)
```

---

### <font color = 'Blue'> 2. Train Test split of dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.33,random_state=10) 
```

```python
print('Length of our dataset is',len(dataset))

print('Length of our train data is',len(x_train))
# Test data is 33% of actual data
print('Length of our test datat is',len(x_test))
```

---

### <font color = 'Blue'> 3.Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_train
```

```python
# we use mean and std of x_train because model should not have any idea about test data
x_test = scaler.transform(x_test)
x_test
```

---

### <font color = 'Blue'> 4. Model training

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()

regressor
```

```python
regressor.fit(x_train,y_train)
```

#### We cannot see the daigram because we have 8 features we cannot see 8D daigram

```python
# we will get 8 different coefficient or slope 
regressor.coef_
```

```python
# we will get 1 intercept
regressor.intercept_
```

---

### <font color = 'Blue'> 5. Prediction

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

---

### <font color = 'Blue'> 6. Calculate the Error 

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
rmse = np.sqrt(mse)

print('Mean Square Error =',round(mse,2))
print('Mean Absolute Error =',round(mae,2))
print('Root Mean Square Error =',round(rmse,2))
```

### <font color = '#AA00FF'> Observation :
* Error is less means model is trained well

---

### <font color = 'Blue'> 7. Accuracy

```python
from sklearn.metrics import r2_score
```

```python
r_square = r2_score(y_test,y_pred)

print('Accuracy =',round(r_square*100,2),'% using r square.')
```

```python
# adj_r2 = 1 - ( (1-r2)(n-1) / (n-1-p) )
n = len(y_test)
p = x_test.shape[1]

adjusted_r_square = 1 - ((1-r_square)*(n-1)/(n-1-p))

print('Accuracy =',round(adjusted_r_square*100,2),'% using adjusted r square.')
```

### <font color = '#AA00FF'> Observation :
* We get 60% accuracy which is less because our data may not be linearly scattered we can improve our model using some techniques.

### <font color = 'Blue'> Pickling:

Python pickle module is used for serialising and de-serialising a Python object structure. Any object in Python can be pickled so that it can be saved on disk. What pickle does is that it serialises the object first before writing it to file. Pickling is a way to convert a python object (list, dict, etc.) into a character stream. The idea is that this character stream contains all the information necessary to reconstruct the object in another python script.

```python
import pickle
```

#### save regressor as pickle file

```python
pickle.dump(regressor,open('model.pkl','wb'))
```

#### load pickle file and predict

```python
model = pickle.load(open('model.pkl','rb'))
```

```python
model.predict(x_test)
```
