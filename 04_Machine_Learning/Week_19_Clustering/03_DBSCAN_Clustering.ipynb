# <font color = 'orange'> DBSCAN Clustering

### Used to cluster the non linear data

---

### <font color = 'Blue'> Import Library, load Dataset

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x, y = make_moons(n_samples = 250, noise = 0.12)
```

### <font color = '#AA00FF'> Observation :
* We will not use the y i.e output point as it's unsupervised learning.

```python
x
```

```python
plt.scatter(x[:,0],x[:,1])
```

### Just to visual and see our 2 category using y as color. we should not use the y as it's unsupervised machine learning.

```python
plt.scatter(x[:,0],x[:,1], c = y)
```

---

### <font color = 'Blue'> Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)

x_scaled
```

---

### <font color = 'Blue'> Clustering

```python
# we need to set our epsilon value i.e radius

dbscan = DBSCAN(eps = 0.3)
```

```python
# aplying to our model
dbscan.fit(x_scaled)
```

```python
dbscan.labels_
```

### <font color = '#AA00FF'> Observation :
* We can notice that we have 0 and 1 that say the clusters and there are few -1 those are the outliers.
* We can observe those outliers in below scatter plot.

```python
plt.scatter(x[:,0],x[:,1],c = dbscan.labels_)
```

### <font color = '#AA00FF'> Observation :
* We can see that DBSCAN can classify the non linear data.
