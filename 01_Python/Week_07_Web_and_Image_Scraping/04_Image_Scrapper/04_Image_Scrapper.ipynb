```python
import requests 
from bs4 import BeautifulSoup
from urllib.request import urlopen
import logging
import os 
```

```python
# Create a folder if not exists to store images
save_dir = 'images/'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
# To avoid blocking of user for frequent hit to scrap the images
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"}
```

```python
query = 'elon musk'
response = requests.get(f"https://www.google.com/search?q={query}&sxsrf=AJOqlzUuff1RXi2mm8I_OqOwT9VjfIDL7w:1676996143273&source=lnms&tbm=isch&sa=X&ved=2ahUKEwiq-qK7gaf9AhXUgVYBHYReAfYQ_AUoA3oECAEQBQ&biw=1920&bih=937&dpr=1#imgrc=1th7VhSesfMJ4M")
print(response)
```

```python
soup = BeautifulSoup(response.content,'html.parser')
print(soup)
```

```python
image_tages = soup.find_all('img')
del image_tages[0]
print(len(image_tages))
print(image_tages)
```

```python
# we want to extract the url of the image
image_tages[0]['src']
```

```python
image_data_mongo = []
for image in image_tages:
    image_url = image['src']
    image_data = requests.get(image_url).content    
    # add the image into list of dictionary 
    my_dict = {'index' : image_url , 'image': image_data}
    image_data_mongo.append(my_dict)    
    # add the image into the image folder
    # image name is query + index of image in the list of image tages
    with open(os.path.join(save_dir,f'{query}_{image_tages.index(image)}.jpg'), 'wb') as f :
        f.write(image_data)
```

```python
# this binary data we can upload in the mongodb
print(image_data_mongo)
```
