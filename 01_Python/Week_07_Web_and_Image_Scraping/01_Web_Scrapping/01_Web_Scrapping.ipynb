```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging
```

```python
flipkart_url = "https://www.flipkart.com/search?q="+"mitv"
print(flipkart_url)
```

```python
# When we hit the url we will get the http response
url_client = urlopen(flipkart_url)
print(url_client)
```

```python
# We can read the html page using httpresponse
flipkart_page = url_client.read()
print(flipkart_page)
```

```python
# it will beautify to make it human understandable
flipkart_html = bs(flipkart_page,'html.parser')
print(flipkart_html)
```

```python
# Now we have to go inside the particular page
big_box = flipkart_html.find_all('div',{'class':'_1AtVbE col-12-12'})
print(len(big_box))
print(big_box)
```

```python
# deleting first 2 and last 2 unwanted div
del big_box[:2]
del big_box[-4:]
print(len(big_box))
```

```python
# Here we have to find the product url to reach the product page
# print(big_box[0].div.div.div.a['href']) -> getting first product link 
products_link = []
for product_overview in big_box:
    products_link.append('https://www.flipkart.com' + product_overview.div.div.div.a['href'])
print(products_link)
```

```python
# We will try to understand how to get the product page
# We will get the response page to read it we have to open in text format
product = requests.get(products_link[0])
print(product)
print(product.text)
```

```python
# Now we will beautify the page to make it human readable format
product_html = bs(product.text,'html.parser')
product_html
```

```python
# Now we have to scrap the review
comment_box = product_html.find_all('div',{'class':'_16PBlm'})
print(len(comment_box))
```

```python
# we have to get the name of the commentor
comment_box[0].div.div.find_all('p',{'class':'_2sc7ZR _2V5EHH'})[0].text
```

```python
# We need to extract the rating of the commentor
comment_box[0].div.div.div.find_all('div',{'class':'_3LWZlK _1BLPMq'})[0].text
```

```python
# We need to extract the comment summary
comment_box[0].div.div.div.p.text
```

```python
# We need to extract the actual comment
comment_box[0].div.div.find_all('div',{'class':''})[1].text
```

```python
# deleting unwanted rows
del comment_box[10:]
```

```python
# now we know how to take a name so we will extract all the name 
commenter_names = []
commenter_rating = []
commenter_highlight = []
commenter_comment = []
for comment in comment_box:
    commenter_names.append(comment.div.div.find_all('p',{'class':'_2sc7ZR _2V5EHH'})[0].text)
    commenter_rating.append(comment.div.div.div.find_all('div',{'class':'_3LWZlK _1BLPMq'})[0].text)
    commenter_highlight.append(comment.div.div.div.p.text)
    commenter_comment.append(comment.div.div.find_all('div',{'class':''})[1].text)
print(commenter_names)
print(commenter_rating)
print(commenter_highlight)
print(commenter_comment)
```
