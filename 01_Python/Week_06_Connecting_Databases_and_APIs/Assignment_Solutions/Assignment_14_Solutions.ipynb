### Q3. What is App routing in Flask? Why do we use app routes?

#### App routing in Flask refers to the process of mapping URLs to specific functions or views within a Flask web application. Flask provides a routing mechanism that allows you to define how different URL paths should be handled and which Python functions should be executed when a particular URL is requested by a client 
Usage :
1. URL Mapping : App routes allow you to map URLs to specific functionality in your web application.
2. View Functions : App routes enable you to associate view functions (Python functions) with specific URLs. 
3. Modularity : Flask's routing system promotes a modular approach to web application development.
4. RESTful Routing : we use GET , POST , PUT , DELETE
5. User-Friendly URLs : You can define user-friendly and human-readable URLs that make it easier for users to navigate your website or access specific resources.
6. Separation of Concerns
