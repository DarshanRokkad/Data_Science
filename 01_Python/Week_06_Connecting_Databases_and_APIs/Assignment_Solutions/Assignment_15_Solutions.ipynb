### Q1. Explain GET and POST methods.

GET Method:

* Used for requesting data.
* Data is in the URL (query parameters).
* Limited data length due to URL constraints.
* Requests can be cached.
* Less secure for sensitive data.
* example : Normal google search.

POST Method:

* Used for submitting data to be processed.
* Data is in the request body.
* Can send larger amounts of data.
* Requests are typically not cached.
* More secure for sensitive data.
* example : Used while logging to some account.

### Q2. Why is request used in Flask?

#### In Flask, the request object is used to access and work with data that a client sends to the server in an HTTP request. It allows your Flask application to retrieve information such as form data, URL parameters, and request headers sent by a user's web browser.

#### In Flask, the redirect() function is used to send an HTTP redirect response to the client's web browser.
