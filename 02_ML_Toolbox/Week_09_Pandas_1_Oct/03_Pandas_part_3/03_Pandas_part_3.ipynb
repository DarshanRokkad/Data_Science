### merge

```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
})
data1
```

```python
data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
})
data2
```

```python
# intersection
pd.merge(data1,data2)
```

```python
# outer join
pd.merge(data1,data2,how='outer')
```

```python
# merge left
pd.merge(data1,data2,how='left')
```

```python
# merge right
pd.merge(data1,data2,how='right')
```

```python
pd.merge(data1,data2,how='cross')
```

---

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2,how='left',left_on='key2',right_on='key4')
```

### join

```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                     index = ['a','b','c','d','e'])
data1
```

```python
data2 = pd.DataFrame({'key11':[1,2,4,5,6],
                      'key22':[4,5,6,7,8],
                      'key33':[3,4,5,6,6]},
                     index=['a','b','h','i','j'])
data2
```

```python
# by default it is left join
data1.join(data2)
```

```python
data1.join(data2,how='right')
```

```python
data1.join(data2,how='inner')
```

```python
data1.join(data2,how='outer')
```

```python
data1.join(data2,how='cross')
```

```python
# increase the fare amount by 10 times

df['Fare_inc'] = df['Fare'].apply(lambda x : x*10)
```

```python
df.head(3)
```

```python
df['Name_len'] = df['Name'].apply(len)
```

```python
df.head(3)
```

```python
# increase Fare by 80 times
def incr_con(x):
    return x*80

df['Fare_inc'] =  df['Fare_inc'].apply(incr_con)
```

```python
df.head(3)
```

```python
def create_flag(x):
    if x<10:
        return 'cheap'
    elif x>=10 and x <20:
        return 'mid'
    else:
        return 'high'
```

```python
df['Flag_fare'] = df['Fare'].apply(create_flag)
```
