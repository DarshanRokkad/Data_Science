```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

---

### Scatter plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
print(x)
print()
print(y)
```

```python
plt.scatter(x,y)
```

```python
# c - color
# alpha - intensity of color 
# marker - shape of the point

# plt.figure(figsize=(6,2))
plt.scatter(x,y,c = 'violet',alpha=0.7 , marker='x')
plt.xlabel('This is my x axis.')
plt.ylabel('This is my y axis.')
plt.title('Scatter plot graph title')
# plt.grid()
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)

print(x)
print()
print(y)
```

```python
# we will get the sine wave
plt.scatter(x,y)
```

---

### Line plot

```python
plt.plot(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,'--r')
plt.xlabel('x value')
plt.ylabel('sinx value')
plt.title('Sine wave')
# plt.grid()
```

---

### Bar plot - bar() and barh()

```python
x = ['a', 'b' , 'c','d','e']
y = np.random.rand(5)

print(x)
print()
print(y)
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y,color='#7E10D5')
plt.xlabel('categorical data')
plt.ylabel('numerical data')
plt.title('Bar plot')
# plt.grid()
```

```python
# barh -> horizontal
plt.figure(figsize=(5,3))
plt.barh(x,y,color='#7E10D5')
plt.xlabel('numerical data')
plt.ylabel('categorical data')
plt.title('Bar plot')
# plt.grid()
```

---

### line plot vs scatter plot

```python
x = [3,4,5,6,7,8]
y = [4,5,6,7,8,9]
```

```python
# line plot -> this will join the points present in the scatter plot
plt.figure(figsize=(4,2))
plt.plot(x,y)
```

```python
# scatter plot -> give the raw plot
plt.figure(figsize=(4,2))
plt.scatter(x,y)
```

---

### Histogram

```python
data = [1,1,2,3,3,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize=(6,4))
plt.hist(data,color='#700EBE')
plt.xlabel('Unique values')
plt.ylabel('Frequency')
plt.title('Histogram')
# plt.grid()

plt.savefig('05_histogram.png')
```

---

### 3D plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z,c='#9327E8',alpha = .9)

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.show()
```
