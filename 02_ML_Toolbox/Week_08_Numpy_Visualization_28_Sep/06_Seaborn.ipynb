```python
import seaborn as sns
```

```python
# seaborn contain some inbuilt dataset so we will use that to learn seaborn
iris = sns.load_dataset('iris')
iris
```

---

### Scatter plot

```python
sns.scatterplot(x = iris.sepal_length , y = iris.sepal_width)
# here there are 2 group 
# sepal length increases then sepal width decrease 
# sepal length increases then sepal width increaes.
```

```python
sns.scatterplot(x = iris.sepal_length , y = iris.petal_length)
# here there are 2 group 
# sepal length increases then petal length not increases. 
# sepal length increases then petal length increases.
```

---

### Histogram

```python
sns.displot(iris['sepal_length'])
```

### tips dataset in seaborn analysics

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x = tips.total_bill , y = tips.tip)
# person who pays more bill pays more tips to worker in hotel
```

```python
tips.head()
```

---

### Relational plot

```python
# pandas group function based on count
tips['smoker'].value_counts()
```

```python
# to finding wheather the person paid more tips is smoker or not smoker
# hue is the color representation
sns.relplot(x = tips.total_bill , y = tips.tip,data = tips,hue = 'smoker')
```

```python
tips['size'].value_counts()
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip , data = tips , hue = 'size')
# but here light and dark color representation is some what different
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip , data = tips , style = 'size')
```

```python
sns.relplot(x = tips.total_bill , y = tips.tip , data = tips , style = 'size', hue = 'time')
```

```python
tips
```

---

### Categorical plot

```python
# only which more total bill is paid
sns.catplot(x = 'day' , y = 'total_bill' ,data = tips)
```

```python
sns.catplot(x = 'smoker' , y = 'tip' ,data = tips)
# more non smoker gives the tip
```

---

### Join plot - scatter + histogram(distplot)

```python
sns.jointplot(x  = tips.total_bill , y = tips.tip)
```
