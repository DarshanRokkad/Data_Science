### <font color='blue'> 1. Drawing a line.

```python
img = cv2.imread('Resources/whiteboard.png')

# image, starting coordinates, ending coordinates, color, thickness
cv2.line(img, (100, 150), (500, 500), (0, 0, 255), 3)

cv2.imshow('line on image', img)
cv2.waitKey(0)
```

#### Output screenshot:


---

### <font color='blue'> 2. Drawing a rectangle.

```python
img = cv2.imread('Resources/whiteboard.png')

# border line rectangle
# image, starting coordinates, ending coordinates, color, thickness
cv2.rectangle(img, (200, 350), (450, 600), (255, 0, 0), 10)

# filled rectangle
# image, starting coordinates, ending coordinates, color, thickness
cv2.rectangle(img, (800, 200), (500, 400), (0, 255, 0), -1)

cv2.imshow('rectangle on image', img)
cv2.waitKey(0)
```

#### Output screenshot:


---

### <font color='blue'> 3. Drawing a circle.

```python
img = cv2.imread('Resources/whiteboard.png')

# image, center, radius, color, thickness
cv2.circle(img, (500, 350), 200, (0, 0, 255), 5)

cv2.imshow('circle on image', img)
cv2.waitKey(0)
```

#### Output screenshot:


---

### <font color='blue'> 4. Drawing a text.

```python
img = cv2.imread('Resources/whiteboard.png')

# image, text, starting, font style, font size, color, thickness
cv2.putText(img, 'Darshan R.M.', (230, 350), cv2.FONT_HERSHEY_SIMPLEX, 2, (120, 80, 130), 5)

cv2.imshow('text on image', img)
cv2.waitKey(0)
```
