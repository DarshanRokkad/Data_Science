## Working with Vedios

### <font color='blue'> Read Video

```python
video = cv2.VideoCapture('data/music_box_vedio.mp4')
```

---

### <font color='blue'> Visualize Video

```python
ret = True 

while ret:
    ret, frame = video.read()    
    if ret:        # true if frame is present
        cv2.imshow('frame', frame)
        cv2.waitKey(40) # wait 40ms
        
video.release()
cv2.destroyAllWindows()
```

---

## Working with Web Cam

### <font color='blue'> Read Webcam

```python
webcam = cv2.VideoCapture(0)
```

### <font color='blue'> Visualize Webcam

```python
while True:
    ret, frame = webcam.read()
    
    cv2.imshow('frame', frame)
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

webcam.release()
cv2.destroyAllWindows()
```
