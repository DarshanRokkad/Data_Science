# <font color = 'orange'> Edge Detection

---

```python
import cv2
```

---

### <font color='blue'> Original image

```python
img = cv2.imread('Resources/bb5.jpeg')                            # shape = (275, 183, 3)
img = cv2.resize(img, (img.shape[1] * 3, img.shape[0] * 3))       # (width, height)

cv2.imshow('image', img)
cv2.waitKey(0)
```

#### Output screenshot:


---

#### Types :
1. Sobel Operator.
2. Laplace Operator.
3. Canny Operator.

---

### <font color='blue'> 3. Canny Operator

```python
edge_img = cv2.Canny(img, 80, 250)          # max value, min value

cv2.imshow('canny operator edge detection', edge_img)
cv2.waitKey(0)
```

#### Output screenshot:


---

###  Dilate - making edges thicker

```python
import numpy as np

edge_img_d = cv2.dilate(edge_img, np.ones((3, 3), dtype = np.int8))          # max value, min value

cv2.imshow('dilute edge image', edge_img_d)
cv2.waitKey(0)
```

#### Output screenshot:


---

###  Erode - making edges thinner

```python
edge_img_e = cv2.erode(edge_img_d, np.ones((3, 3), dtype = np.int8))          # max value, min value

cv2.imshow('Erode edge image', edge_img_e)
cv2.waitKey(0)
```
