```python
import cv2
```

---

### <font color='blue'> Image import

```python
img = cv2.imread('Resources/paroat1.jpeg')

cv2.imshow('image', img)
cv2.waitKey(0)
```

* Image normally will be in BGR color.

---

### <font color='blue'> Image color convertion

```python
# bgr to rgb color conversation
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cv2.imshow('bgr image', img)
cv2.imshow('rgb image', rgb_img)

cv2.waitKey(0)
```

### Output screenshot :


```python
# gray scale image

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('Gray Image', gray_img)
cv2.waitKey(0)
```

### Output screenshot :


```python
# hsv image
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

cv2.imshow('HSV Image', hsv_img)
cv2.waitKey(0)
```
