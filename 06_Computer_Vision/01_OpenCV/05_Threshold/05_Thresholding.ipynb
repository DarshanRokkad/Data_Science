```python
import cv2
```

### <font color='blue'> Original image

```python
img = cv2.imread('Resources/bear.jpg')

cv2.imshow('image', img)
cv2.waitKey(0)
```

#### Output screenshot:

### <font color='blue'> Convert to gray scale image

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('Gray Scale Image', gray_img)
cv2.waitKey()
```

#### Output screenshot:


---

### Types
1. Simple Thresholding.
2. Adaptive Thresholding.

---

### <font color='blue'> 1. Simple Thresholding

```python
ret, thres_img = cv2.threshold(gray_img, 80, 255, cv2.THRESH_BINARY)

cv2.imshow('threshold image', thres_img)
cv2.waitKey(0)
```

#### Output screenshot:


---

```python
# improving thresholding

# just adding a blur to remove noise present in image
ret, thres_img = cv2.threshold(gray_img, 80, 255, cv2.THRESH_BINARY)

blur_img = cv2.medianBlur(thres_img, 5)

cv2.imshow('blur threshod image', blur_img)
cv2.waitKey(0)
```

#### Output screenshot:


---

### <font color='blue'> 2. Adaptive Thresholding

```python
# original image
img = cv2.imread('Resources/handwritten.png')

cv2.imshow('original image', img)
cv2.waitKey(0)

# in image as we observe there is different shade so we need threshold which will adapt this changes and give image correctly 
# so we can apply 1 single threshold that we can apply 
```

#### Output screenshot:


---

```python
# simple thresholding
ret, thres_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

cv2.imshow('threshold image', thres_img)
cv2.waitKey(0)
```

#### Output screenshot:


---

```python
# adaptive thresholding

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

adap_thres_1 = cv2.adaptiveThreshold(gray_img , 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 21, 30)
adap_thres_2 = cv2.adaptiveThreshold(gray_img , 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 30)

cv2.imshow('Adaptive Mean Thresholding', adap_thres_1)
cv2.imshow('Adaptive Gaussian Thresholding', adap_thres_2)

cv2.waitKey(0)
```
