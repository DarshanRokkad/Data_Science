## Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

T-test and Z-test are both statistical hypothesis tests used to make inferences about population parameters based on sample data.  
However, they are used under different circumstances and have some key differences:

1. **Population Variance Known vs. Unknown**:
    * **Z-test**: It is used when you know the population's standard deviation (σ). It's suitable for large sample sizes when you can reasonably assume the population standard deviation is known.
    * **T-test**: It is used when the population standard deviation is unknown, and you estimate it from the sample (s). It is appropriate for smaller sample sizes.

2. **Sample Size**:
    * **Z-test**: Generally used with large sample sizes (typically n > 30) because it relies on the approximation of the normal distribution.
    * **T-test**: Designed for smaller sample sizes, and it's more robust in such cases.

3. **Test Statistic**:
    * **Z-test**: Uses the Z-statistic, which follows the standard normal distribution (mean of 0 and standard deviation of 1).
    * **T-test**: Uses the T-statistic, which follows a Student's t-distribution with varying degrees of freedom.

Here's an example scenario for each test:

**Z-test**:
Suppose you are a quality control manager at a cookie manufacturing company, and you want to test if the mean weight of a specific cookie type is consistent with the industry-standard weight of 15 grams. You have a large production batch, and you know the population standard deviation of the cookie weights is 2 grams (σ = 2). In this case, you would use a Z-test to determine if the mean weight of the cookies in your sample is significantly different from 15 grams.

**T-test**:
Imagine you are a researcher conducting a study to compare the test scores of two groups of students, but you don't know the population standard deviation of the scores. You collect a sample of 20 students from each group and calculate the sample standard deviations. In this scenario, you would use a two-sample T-test to assess whether there is a statistically significant difference in the means of the two groups' test scores.

## Q2: Differentiate between one-tailed and two-tailed tests.

One-tailed and two-tailed tests are two types of statistical hypothesis tests used to determine whether an observed sample statistic falls within a certain range of values or if it significantly differs from a specific value or range.  
The key difference between them lies in the directionality of the hypothesis being tested and the critical region.

**One-Tailed Test**:
1. **Directionality**: A one-tailed test, also known as a one-sided test, is used when you have a specific expectation about the direction of the effect or difference in your hypothesis. In other words, you are interested in testing if the population parameter is greater than or less than a particular value, but not both.

2. **Critical Region**: In a one-tailed test, the critical region (the region of extreme values) is located on only one side of the probability distribution. It's either in the right tail (for testing if the parameter is greater) or the left tail (for testing if the parameter is less).

3. **Hypotheses**:
    * For a **Greater Than Test** (right-tailed): 
        * Null Hypothesis (H0): μ (population parameter) ≤ a specified value
        * Alternative Hypothesis (Ha): μ > a specified value

    * For a **Less Than Test** (left-tailed):
        * Null Hypothesis (H0): μ ≥ a specified value
        * Alternative Hypothesis (Ha): μ < a specified value

**Two-Tailed Test**:
1. **Directionality**: A two-tailed test is used when you are interested in testing if the population parameter is significantly different from a specific value, but you do not have a specific expectation about the direction of the effect. In other words, you want to detect deviations in both directions from the specified value.

2. **Critical Region**: In a two-tailed test, the critical region is divided into two parts, one in each tail of the probability distribution.

3. **Hypotheses**:
    * Null Hypothesis (H0): μ (population parameter) = a specified value
    * Alternative Hypothesis (Ha): μ ≠ a specified value

**Example**:

Suppose you are testing a new drug's effect on blood pressure. You want to determine if the drug either lowers or raises blood pressure, but you don't have a specific expectation about the direction. 

* For a **Two-Tailed Test**:
    * Null Hypothesis (H0): The drug has no effect on blood pressure (μ = 0).
    * Alternative Hypothesis (Ha): The drug has an effect on blood pressure (μ ≠ 0).

* For a **One-Tailed Test**:
    * Null Hypothesis (H0): The drug has no effect on blood pressure (μ = 0).
    * Alternative Hypothesis (Ha): 
        * If you expect the drug to lower blood pressure, the right-tailed test would be: μ < 0.
        * If you expect the drug to raise blood pressure, the left-tailed test would be: μ > 0.

## Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

Type 1 and Type 2 errors are fundamental concepts in hypothesis testing, which is a statistical method used to make decisions about population parameters based on sample data. These errors are associated with the decisions made when testing a null hypothesis.

1. **Type 1 Error** (False Positive):
    Type 1 error occurs when you reject a null hypothesis that is actually true. In other words, it is a false positive or a "false alarm."

   Example Scenario for Type 1 Error:
   Suppose a pharmaceutical company is testing a new drug to determine if it is effective in treating a certain medical condition. The null hypothesis (H0) states that the drug has no effect, while the alternative hypothesis (Ha) suggests that the drug is effective. If the company sets the significance level at α = 0.05, and during their study, they incorrectly reject the null hypothesis (i.e., they conclude that the drug is effective) when, in reality, it has no effect, they would be making a Type 1 error.  
   
   
2. **Type 2 Error** (False Negative):
    Type 2 error occurs when you fail to reject a null hypothesis that is actually false. In other words, it is a false negative or a "miss."

   Example Scenario for Type 2 Error:
   Consider a quality control scenario in a factory. The null hypothesis (H0) states that the manufacturing process is producing products within acceptable quality standards, while the alternative hypothesis (Ha) suggests that the process is producing subpar products. If the factory uses a significance level of α = 0.05 and fails to reject the null hypothesis, even though the process is producing subpar products, they would be making a Type 2 error.

In summary:

* Type 1 error is when you incorrectly reject a true null hypothesis (false positive).
* Type 2 error is when you fail to reject a false null hypothesis (false negative).

## Q4:  Explain Bayes's theorem with an example.

Bayes's theorem is a fundamental concept in probability theory and statistics that describes **how to update the probability of an event** based on new evidence or information.  
It provides a way to revise our beliefs about the probability of an event as we receive additional data.  
The theorem is named after Thomas Bayes, an 18th-century mathematician.

The formula for Bayes's theorem is as follows:

P(A|B) = ( P(A) x P(B|A) ) / P(B)

Where:
* P(A|B) is the conditional probability of event A occurring given that event B has occurred.
* P(B|A) is the conditional probability of event B occurring given that event A has occurred.
* P(A) is the prior probability of event A occurring.
* P(B) is the prior probability of event B occurring.

Bayes's theorem allows us to **update our initial belief (prior probability) about an event (A)** in light of **new evidence (B)** to calculate the revised probability (posterior probability).

Let's illustrate Bayes's theorem with a practical example:

**Example: Medical Diagnosis**

Suppose you are a doctor trying to diagnose whether a patient has a particular disease, which we'll call "Disease X." You have the following information:

1. The prior probability of a patient having Disease X is (P(Disease, X) = 0.01) (1% of the population has the disease).
2. You have a diagnostic test for Disease X that is not perfect:
    * If a patient has the disease, the test correctly identifies it 95% of the time: (P(Positive Test | Disease X) = 0.95).
    * If a patient does not have the disease, the test gives a false positive 3% of the time: (P(Positive Test | No Disease X) = 0.03).

Now, a patient comes in and takes the test, and the test comes back positive. You want to know the probability that the patient actually has Disease X.

Using Bayes's theorem, we can calculate this:

1. **Prior Probability**: (P(Disease X) = 0.01)
2. **Prior Probability of No Disease X**: (P(No Disease X) = 1 - P(Disease X) = 0.99)
3. **Likelihood of a Positive Test Given Disease X**: (P(Positive Test | Disease X) = 0.95)
4. **Likelihood of a Positive Test Given No Disease X**: (P(Positive Test | No Disease X) = 0.03)

Now, we can apply Bayes's theorem:

P(Disease X | Positive Test) = (P(Positive Test|Disease X) x P(Disease X)) / P(Positive Test)

To calculate P(Positive Test) we can use the law of total probability:

P(Positive Test) = (P(Positive Test | Disease X) / P(Disease X)) + (P(Positive Test | No Disease X) / P(No Disease X))

Substitute the values and calculate to find the revised probability that the patient has Disease X given the positive test result.

Bayes's theorem allows us to make more informed decisions and **update our beliefs** as we receive new data, making it a powerful tool in various fields, including medicine, machine learning, and more.

## Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

A confidence interval is a statistical range or interval that is used to estimate a population parameter (such as a population mean or proportion) based on a sample of data.  
It provides a way to express the uncertainty or margin of error associated with the estimate.   
The confidence interval represents a range of values within which we believe the true population parameter is likely to fall with a certain level of confidence.

Confidence intervals are typically expressed with two values: a point estimate (usually the sample statistic) and a margin of error. The level of confidence is expressed as a percentage (commonly 95% or 99%), and it quantifies the probability that the interval contains the true population parameter.

To **calculate a confidence interval**, we have to follow the following steps:

1. A sample from the population.
2. The sample statistic (e.g., sample mean or sample proportion).
3. The standard error of the statistic.
4. The desired level of confidence (e.g., 95% confidence level).

**Example: Confidence Interval for Population Mean**

Suppose you want to estimate the average height of adult males in a particular city. You collect a random sample of 50 adult males and measure their heights. The sample mean height is 175 cm, and you want to calculate a 95% confidence interval for the **population mean height**.

The steps to calculate the confidence interval are as follows:
1. **Calculate the sample mean and sample standard deviation**:
    * Sample mean (x bar): In this case, it's 175 cm.
    * Sample standard deviation (s): Let's say it's 8 cm.  
    
    
2. **Determine the desired confidence level**:
    * You want a 95% confidence interval.  
    
    
3. **Find the critical value for the chosen confidence level**:
    * For a 95% confidence level, the critical value is typically obtained from a standard normal distribution or a t-distribution (depending on the sample size and whether the population standard deviation is known). Let's assume you're using a z-distribution for simplicity, and the critical value for a 95% confidence interval is approximately 1.96.  
    
    
4. **Calculate the standard error**:
    * Standard error (\(SE\)) is calculated as \( \frac{s}{\sqrt{n}} \), where \( n \) is the sample size. In this case, \( SE = \frac{8}{\sqrt{50}} \).  
    
    
5. **Calculate the margin of error**:
    * Margin of error (\(ME\)) is calculated by multiplying the standard error by the critical value: \( ME = 1.96 \times \frac{8}{\sqrt{50}} \).  
    
    
6. **Calculate the confidence interval**:
    * Lower limit of the confidence interval: \( \bar{x} - ME = 175 - (1.96 \times \frac{8}{\sqrt{50}}) \).
    * Upper limit of the confidence interval: \( \bar{x} + ME = 175 + (1.96 \times \frac{8}{\sqrt{50}}) \).  
    
    
In this example, the 95% confidence interval for the average height of adult males in the city is calculated to be, for instance, 172.96 cm to 177.04 cm. This means that you can be 95% confident that the true average height of adult males in the city falls within this interval.

## Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

The margin of error (MOE) in a confidence interval is a measure of the precision or the **amount of uncertainty associated with the estimate of a population parameter**, such as a population mean or proportion.  
It represents the range within which we believe the true parameter is likely to fall with a certain level of confidence.  
A smaller margin of error indicates a more precise estimate, while a larger margin of error indicates more uncertainty.

The **margin of error is typically affected** by three main factors:

1. **Confidence Level**: The level of confidence you choose (e.g., 95% or 99%) affects the margin of error. **Higher confidence levels** result in wider intervals and **larger margins of error**, as you are more certain that the true parameter lies within the interval.  


2. **Standard Deviation or Standard Error**: The standard deviation (or standard error) of the population or sample being measured affects the margin of error. A **larger standard deviation** leads to a **larger margin of error**, as it indicates greater variability in the data.  


3. **Sample Size**: Sample size has a significant impact on the margin of error. A **larger sample size** generally results in a **smaller margin of error** because it provides more information about the population. Larger samples tend to produce more precise estimates of the population parameter.

**Example: Impact of Sample Size on Margin of Error**

Suppose you want to estimate the average income of households in a city. You collect data from two different sample sizes:

* Sample A: 100 households
* Sample B: 1,000 households

Assuming the same level of confidence and a similar standard deviation for both samples, you'll find that **Sample B will have a smaller margin of error compared to Sample A**. Here's why:

* Sample B is larger, providing more information about the population. With **more data points**, you have a **better estimate of the population's true average income**.
* The margin of error : **Sample B** will be **smaller** because it is **more precise**
* The margin of error : **Sample A** will be **larger** because it has **less information**, resulting in greater uncertainty.
