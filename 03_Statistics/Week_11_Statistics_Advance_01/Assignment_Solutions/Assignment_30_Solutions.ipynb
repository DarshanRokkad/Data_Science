## Q1. What is Estimation Statistics? Explain point estimate and interval estimate.

Estimation statistics involves making estimation about unknown population parameters based on sample data.  
There are two main types of estimates:

1. **Point Estimate**: A **single value** that's the best guess for the population parameter based on the sample data.  
    Example : Imagine you want to estimate the average income of people in a city. You randomly survey 100 residents and find that their average income is dollar 50,000 per year. In this case, dollar 50,000 is the point estimate of the average income for the entire city's population.  
    
    
2. **Interval Estimate**: A range of values within which the true population parameter is likely to fall. It considers both point estimation and uncertainty i.e margin of error, often expressed as a **confidence interval** (e.g., 95% confidence), indicating the level of confidence in the estimate.  
    Example : Using the same example, if you calculate a 95% confidence interval for the average income, you might find that it ranges from dollar 48,000 to dollar 52,000. This means you are 95% confident that the true average income of the city's population falls within this range. So, you have both a point estimate (dollar 50,000) and an interval estimate (dollar 48,000 to dollar 52,000) that accounts for the uncertainty in your estimate.

## Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.

```python
import math 
def estimate_population_mean(sample_size , sample_mean , sample_std):
    # confidence interval =  x̄ ± Z * (s / √n)
    
    standard_error = sample_std / math.sqrt(sample_size)
    # z score for 95% confidence interval or 5% significance level 
    z_score = 1.96
    margin_of_error = z_score * standard_error
    
    lower_estimate = sample_mean - margin_of_error
    upper_estimate = sample_mean + margin_of_error    
    confidence_interval = [lower_estimate , upper_estimate]
    
    return lower_estimate , upper_estimate

sample_size = 30 
sample_mean = 100  
sample_std = 15  

lower_estimate,upper_estimate = estimate_population_mean(sample_size,sample_mean, sample_std)

print(f"Estimated Population Mean: {sample_mean}")
print(f"95% Confidence Interval: ({lower_estimate}, {upper_estimate})")
```

## Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

```python
# t - test

import numpy as np
from scipy import stats

sample1 = [68, 72, 74, 71, 68, 70, 73, 72, 69, 75]
sample2 = [63, 66, 61, 65, 62, 64, 68, 67, 63, 68]

# Perform the independent two-sample t-test
t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

# sigificance value
alpha = 0.05

print("Sample 1 Mean:", np.mean(sample1))
print("Sample 2 Mean:", np.mean(sample2))
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the means.")
```

## Q6. What is a null and alternative hypothesis? Give some examples.

In statistical hypothesis testing, the null hypothesis (H0) and the alternative hypothesis (H1 or Ha) are two complementary statements used to make inferences about an unknown population parameter based on sample data.  

1. **Null Hypothesis (H0):**
    * The null hypothesis is the default or baseline assumption that there is no significant effect, relationship, or difference in the population. It represents a lack of change or an absence of an effect.
    * The null hypothesis is what you aim to test against to determine whether there's enough evidence to reject it.
    * In notation, H0 typically includes statements like "equal to," "no difference," or "no effect."

   Examples of null hypotheses:
    * The average height of male and female students in a school is equal (H0: μ_male = μ_female).
    * A new drug has no effect on reducing blood pressure compared to a placebo (H0: μ_drug = μ_placebo).
    * There is no association between smoking and the risk of lung cancer (H0: There is no association).  
      
      
2. **Alternative Hypothesis (H1 or Ha):**
    * The alternative hypothesis is the statement you want to support or prove, indicating that there is a significant effect, relationship, or difference in the population.
    * It represents the opposite of the null hypothesis and can take several forms, such as "greater than," "less than," or "not equal to."

   Examples of alternative hypotheses corresponding to the above null hypotheses:
    * For the null hypothesis "The average height of male and female students in a school is equal," the alternative hypothesis might be "The average height of male and female students in a school is not equal" (H1: μ_male ≠ μ_female).
    * For the null hypothesis "A new drug has no effect on reducing blood pressure compared to a placebo," the alternative hypothesis might be "The new drug has a greater effect on reducing blood pressure than the placebo" (H1: μ_drug > μ_placebo).

## Q7. Write down the steps involved in hypothesis testing.

Hypothesis testing is a structured process used in statistics to determine whether there is enough evidence in a sample to support a specific hypothesis about a population parameter. The following are the steps typically involved in hypothesis testing:  
  
  
1. **Formulate the Hypotheses**:
    * State the null hypothesis (H0), which represents the default assumption or no effect, and the alternative hypothesis (H1 or Ha), which represents the statement you want to test.
    * Define the population parameter you are investigating and specify the type of relationship (equality, inequality, or association).  
  
  
2. **Select the Significance Level (α)**:
    * Choose a significance level (α) that represents the acceptable level of risk for making a Type I error (false positive). Common values are 0.05 or 0.01, corresponding to 5% and 1% significance levels, respectively.  
  
  
3. **Collect and Analyze Data**:
    * Collect a sample from the population of interest.
    * Calculate sample statistics (e.g., sample mean, standard deviation) and test statistics based on the data.  
  
  
4. **Select a Statistical Test**:
    * Choose an appropriate statistical test based on the type of data, research question, and hypotheses. Common tests include t-tests, chi-square tests, ANOVA, regression analysis, and more.  
  
  
5. **Calculate the Test Statistic**:
    * Compute the test statistic based on the selected test and the sample data. The test statistic measures the difference between the sample data and what is expected under the null hypothesis.  
  
  
6. **Determine the Critical Region**:
    * Determine the critical region, which is the range of values of the test statistic beyond which you would reject the null hypothesis.
   - The critical region is determined based on the chosen significance level (α) and the distribution of the test statistic.  
  
  
7. **Compare the Test Statistic and Critical Region**:
    * Compare the calculated test statistic with the critical region.
    * If the test statistic falls within the critical region, reject the null hypothesis. If it falls outside the critical region, fail to reject the null hypothesis.  
  
  
8. **Calculate the P-Value (Optional)**:
    * If the p-value is less than α, reject the null hypothesis.  
  
  
9. **Make a Decision**:
    * Based on the comparison of the test statistic or p-value with the significance level, make a decision to either reject the null hypothesis or fail to reject it.  
  
  
10. **Draw Conclusions**:
    * Based on the decision made in the previous step, draw conclusions about the population parameter and the hypothesis being tested.  
  
  
11. **Report Results**:
    * Present the results of the hypothesis test, including the test statistic, p-value (if calculated), the decision made, and the conclusions drawn.
    * Be clear about the practical significance of the findings.  
  
  
12. **Interpret the Results**:
    * Interpret the results in the context of the research question and the broader implications for the field of study.

## Q8. Define p-value and explain its significance in hypothesis testing.

The **p-value** (probability value) is a crucial concept in hypothesis testing. It quantifies the **strength of the evidence against the null hypothesis (H0)** and helps researchers or analysts make decisions regarding hypothesis testing. 

1. **Definition**:
    * The p-value is a numerical measure that represents the probability of obtaining a test statistic as extreme as, or more extreme than, the one observed in the sample data, assuming that the null hypothesis is true.
    * In simpler terms, the p-value tells you how likely it is to observe the sample data if the null hypothesis is correct.  
      
      
2. **Significance in Hypothesis Testing**:

    * **Decision Rule**: In hypothesis testing, you compare the p-value to the chosen significance level (α) to make a decision about the null hypothesis. The decision rule is typically as follows:
        * If the **p-value is less than or equal to α, you reject the null hypothesis**.
        * If the p-value is greater than α, you fail to reject the null hypothesis.  
        
    * **Interpretation**:
        * A **small p-value** (typically less than α) indicates that the observed **data is inconsistent** with what would be expected under the null hypothesis. In other words, it suggests that there is strong evidence against the null hypothesis.
        * A **large p-value** (greater than α) suggests that the observed **data is consistent** with what would be expected under the null hypothesis. It does not provide strong evidence against the null hypothesis.  
        
    * **Level of Significance**: The choice of the significance level (α) determines the threshold for what you consider a "small" p-value. Common values for α are 0.05 (5%) and 0.01 (1%). If you set α to 0.05, you are saying that you are willing to accept a 5% chance of making a Type I error (incorrectly rejecting the null hypothesis when it is true).  
    
    * **Practical Significance**: It's important to note that a small p-value does not necessarily mean that the effect or difference being tested is practically significant. Researchers must consider both statistical significance (as indicated by the p-value) and the practical importance or real-world relevance of the findings.  
    
    * **Continuous Information**: The p-value provides continuous information about the strength of the evidence, allowing you to make more nuanced decisions. For example, you can compare p-values across different studies to assess the consistency of findings.
    
    * **Scientific Rigor**: The p-value serves as a rigorous tool for scientific inquiry and decision-making. It helps researchers avoid making arbitrary decisions based on subjective judgment and instead relies on objective statistical evidence.

## Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

degrees_of_freedom = 10

x = np.linspace(-5, 5, 1000)
y = t.pdf(x, df=degrees_of_freedom)

plt.plot(x, y, label=f't-distribution (df = {degrees_of_freedom})', color='b')

plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title("Student's t-Distribution")

plt.legend()
plt.grid(True)

plt.show()
```

## Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.

```python
import numpy as np
from scipy import stats

sample1 = np.random.normal(loc=70, scale=10, size=30)
sample2 = np.random.normal(loc=75, scale=12, size=30)

# Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=True)

# Set the significance level (alpha)
alpha = 0.05

print("Sample 1 Mean:", np.mean(sample1))
print("Sample 2 Mean:", np.mean(sample2))
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the population means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the population means.")
```

## Q11. What is Student’s t distribution? When to use the t-Distribution.

Student's t-distribution, often simply referred to as the t-distribution, is a probability distribution used in statistics. It is similar to the standard normal distribution but is characterized by heavier tails. The t-distribution is particularly useful when dealing with small sample sizes, where sample means may not follow a normal distribution.

Key characteristics of the t-distribution include:

1. **Shape**: The t-distribution has a bell-shaped curve like the normal distribution, but it has heavier tails. This means that the probability of observing extreme values (far from the mean) is higher in a t-distribution compared to a normal distribution.

2. **Parameter**: The shape of the t-distribution depends on the degrees of freedom (df), which is typically associated with sample size. As the degrees of freedom increase, the t-distribution approaches the shape of a standard normal distribution (the normal distribution with a mean of 0 and standard deviation of 1).

3. **Center and Spread**: The t-distribution is centered at 0, and its spread (standard deviation) depends on the degrees of freedom. With a smaller df, the spread is larger, reflecting the increased uncertainty associated with smaller samples.

When to use the t-distribution:

1. **Small Sample Sizes**: The t-distribution is especially useful when working with small sample sizes (typically when the sample size is less than 30). In these cases, sample means may not be normally distributed, and the t-distribution provides a more appropriate model for making inferences about population parameters.

2. **Unknown Population Standard Deviation**: If you don't know the population standard deviation and need to estimate it from the sample, you would use the t-distribution for hypothesis testing and confidence intervals.

3. **Hypothesis Testing**: When conducting hypothesis tests involving sample means, the t-distribution is used when dealing with small samples, allowing for more accurate testing of hypotheses.

4. **Confidence Intervals**: When estimating population parameters (e.g., the mean) and constructing confidence intervals, the t-distribution is used to account for the increased uncertainty associated with smaller sample sizes.

In summary, the t-distribution is a critical tool in statistics for making inferences about population parameters when dealing with small sample sizes or situations where the population standard deviation is unknown.
