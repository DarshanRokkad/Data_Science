# <font color = 'orange'> QQ Plot - Quantile plot

* To check wheather the distribution is normal/gaussian or not QQ plot is used.

---

```python
import scipy.stats as stat
import pylab
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_df = sns.load_dataset('iris')
iris_df
```

---

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde=True)
    
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    
    plt.show()
```

```python
plot_data(iris_df,'sepal_width')
# left side graph basically tells that it is a normal distribution
# right side QQ-plot graph tells wheather distribution is normal distribution or not 
```

```python
plot_data(iris_df,'sepal_length')
```

```python
plot_data(iris_df,'petal_length')
```

```python
plot_data(iris_df,'petal_width')
```
